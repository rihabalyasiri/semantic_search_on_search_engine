# Scatter Plot for "Wein Kaufen" with Experiment no. 3 and Cosine Similarity

```python
import json
import pandas as pd

# Path to your JSON file
file_path = './results/wein_kaufen/top_1425/Experiment_cos_sim_1.json'

# Load JSON data from file
with open(file_path, 'r') as file:
    data1 = json.load(file)

```

```python
df = pd.DataFrame(data1)
filtered_df = df[(df['similarity'] > 0.7) & (df['wein_kaufen'])]
filtered_df
```

```python
import matplotlib.pyplot as plt1
import random


# Extracting data for the plot
x_values = []  # This will store cosine similarity values
y_values = []  # This will store random values for scattering
colors = []    # This will store color values

max_similarity = 0

for obj in data1:
    cosine_similarity = obj.get('similarity', 0)
    wein_kaufen = obj['wein_kaufen']
    max_similarity = max(max_similarity, cosine_similarity) 
    x_values.append(cosine_similarity)
    y_values.append(random.uniform(0, 1) if wein_kaufen else random.uniform(-1, 0))  # Random values for true/false
    colors.append('green' if wein_kaufen else 'blue')  # Yellow for "wein_kaufen" = true, blue for false

# Add the specific query with a large red circle
x_query = [max_similarity]  # Maximum similarity value
y_query = [random.uniform(0, 1)]

# Creating the scatter plot
plt1.figure(figsize=(10, 6))
plt1.scatter(x_values, y_values, c=colors, label='relevant Ads')
plt1.scatter(x_query, y_query, c='red', s=100, label='Query "wein kaufen"')  # Larger red circle for the specific query
plt1.xlabel('Cosine Similarity')
plt1.ylabel('Random Scatter (Wein Kaufen)')
plt1.title('Scatter Plot for "Wein Kaufen" Query')
plt1.legend()
plt1.grid(True)
#plt1.show()
```

# Scatter Plot for "Wein Kaufen" with Experiment no. 3 and Faiss Similarity

```python
import json

# Path to your JSON file
file_path = './results/wein_kaufen/top_1425/Experiment_faiss_similarity.json'

# Load JSON data from file
with open(file_path, 'r') as file:
    data2 = json.load(file)
```

```python
import matplotlib.pyplot as plt2
import random


# Extracting data for the plot
x_wein_kaufen = []  # Distances for "wein_kaufen" = true
x_other_ads = []  # Distances for "wein_kaufen" = false
y_wein_kaufen = []  # Y-values for "wein_kaufen" = true (for scattering)
y_other_ads = []    # This will store color values

min_distance = float('inf')

for obj in data2:
    distance = obj.get('similarity', 0)
    min_distance = min(min_distance, distance)
    wein_kaufen = obj['wein_kaufen']

    if wein_kaufen:
        x_wein_kaufen.append(distance)
        y_wein_kaufen.append(random.uniform(0, 1))  # Random scatter close to 0
    else:
        x_other_ads.append(distance)
        y_other_ads.append(random.uniform(-1, 0))

# Add the specific query
x_query = [min_distance]
y_query = [random.uniform(0, 1)]  # Place it centrally on the Y-axis

# Plotting
plt2.figure(figsize=(10, 6))
plt2.scatter(x_wein_kaufen, y_wein_kaufen, color='green', label='relevant Ads')
plt2.scatter(x_other_ads, y_other_ads, color='blue', label='not relevant Ads')
plt2.scatter(x_query, y_query, color='red', s=100, label='Query "Wein Kaufen"', marker='*')  # Star marker for the query

plt2.xlabel('Euclidean Distance')
plt2.ylabel('Random Scatter')
plt2.title('Scatter Plot for "Wein Kaufen" Query')
plt2.legend()
plt2.grid(True)
plt2.show()
```

```python
import requests
from bs4 import BeautifulSoup

def get_meta_description(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Will raise an HTTPError if the HTTP request returned an unsuccessful status code

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the meta description tag
        meta_description_tag = soup.find('meta', attrs={'name': 'description'})
        
        if meta_description_tag:
            return meta_description_tag.get('content')
        else:
            return "No description meta tag found"
    except requests.RequestException as e:
        return str(e)

# URL to scrape
url = 'https://immobilie-richtig-verkaufen.ch'

# Get and print the meta description
description = get_meta_description(url)
print(description)
```

```python
def plot_grouped_frame( f, labelrotation=15 ):
    """"""
    # setup the plot with two plots in one row
    fig, (ax1,ax2) = plt.subplots( 1,2 , sharey=True, figsize=(12, 3) )
    # plot Relevant@25 on the left
    f.unstack()["Relevant no. of 25"].plot( kind="bar", ax=ax1, ylabel="Relevant", xlabel="", legend=False)
    # plot Relevant@10 on the right
    f.unstack()["Relevant no. of 10"].plot( kind="bar", ax=ax2, ylabel="Relevant", xlabel="" )
    
    # set the titles for each plot
    ax1.set_title("Relevant@25")
    ax2.set_title("Relevant@10")
    
    # set some properties for each of the plots
    for ax in [ax1,ax2]:
        if labelrotation is not None:
            # rotate x-axis labels
            ax.tick_params(axis='x', labelrotation=labelrotation)
        # set gridlines on y-axis
        ax.yaxis.grid(True, linestyle='--', )
        # put axis into back, so that bars are on top
        ax.set_axisbelow(True)
        # remove the spines (box/frame) for the plot
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
```

```python
def code_feature_set( f ):
    """"""
    if "body_text" == f:
        return "Feature Set 1"
    elif "body_text, advertiser domain name" == f:
        return "Feature Set 2"
    elif "body_text, advertiser domain name, Image Caption" == f:
        return "Feature Set 3"
    elif "body_text, advertiser domain name, Image Caption, keywords" == f:
        return "Feature Set 4"
    elif "body_text, advertiser domain name, Image Caption, keywords, combined_description" == f:
        return "Feature Set 5"
    else:
        return ""

# this translates the features (list per cell) into categorical values
df["Feature Set"] = df_cosine["Features"].apply( code_feature_set )
```
